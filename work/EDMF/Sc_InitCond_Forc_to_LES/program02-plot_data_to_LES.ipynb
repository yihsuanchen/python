```python
import matplotlib.pyplot as plt
import numpy as np
import xarray as xr
import io, os, sys, types
import yhc_module as yhc
from datetime import date

xr.set_options(keep_attrs=True)  # keep attributes after xarray operation
today = date.today()
```

## Open data

```python
datapath = "./"

da_scm = xr.open_dataset("./data-DYCOMS_SCM_InitCond.nc")
da_merra2 = xr.open_dataset("./data-MERRA2_InitCond_Forc.nc")
da_s11_ctl = xr.open_dataset("./data-CGILS_S11_ctl_InitCond_Forc.nc")
da_s12_ctl = xr.open_dataset("./data-CGILS_S12_ctl_InitCond_Forc.nc")
```

## Plot initial conditions

```python
#--- set default settings of the figures 
def ax_def (ax, var):
    ax.grid(True)
    ax.minorticks_on()
    ax.invert_yaxis()
    ax.legend(["DYCOMS SCM", "MERRA-2", "CGILS, S11_ctl", "CGILS, S12_ctl"])
    ax.set_title(var.attrs['long_name'], loc='left')
    ax.set_title(var.attrs['units'], loc='right')
    ax.set_xlabel(var.attrs['long_name']+" ("+var.attrs['units']+")")
    ax.set_ylabel("Pressure (hPa)")

#--- open fig
fig, ((ax_U, ax_V, ax_Omega), (ax_T, ax_Q, ax_dum)) = plt.subplots(nrows=2, ncols=3, figsize=(18, 12))
fig.suptitle("Initial Conditions", fontsize=20)

#--- set line styles
ln_scm, ln_merra2, ln_s11_ctl, ln_s12_ctl = 'r-', 'b-', 'g-', 'y-'

#--- plot U
ax_U.plot(da_scm.U, da_scm.plev/100., ln_scm,
          da_merra2.U, da_merra2.plev/100., ln_merra2,
          da_s11_ctl.U, da_s11_ctl.plev/100., ln_s11_ctl,
          da_s12_ctl.U, da_s12_ctl.plev/100., ln_s12_ctl,
         )
ax_def(ax_U, da_scm.U)

#--- plot V
ax_V.plot(da_scm.V, da_scm.plev/100., ln_scm,
          da_merra2.V, da_merra2.plev/100., ln_merra2,
          da_s11_ctl.V, da_s11_ctl.plev/100., ln_s11_ctl,
          da_s12_ctl.V, da_s12_ctl.plev/100., ln_s12_ctl,
         )
ax_def(ax_V, da_scm.V)

#--- plot Omega
factor = 864.
ax_Omega.plot(da_scm.Omega*factor, da_scm.plev/100., ln_scm,
          da_merra2.Omega*factor, da_merra2.plev/100., ln_merra2,
          da_s11_ctl.Omega*factor, da_s11_ctl.plev/100., ln_s11_ctl,
          da_s12_ctl.Omega*factor, da_s12_ctl.plev/100., ln_s12_ctl,
         )
var_dum = da_scm.Omega.copy()
var_dum.attrs['units'] = "hPa/day"
ax_def(ax_Omega, var_dum)

#--- plot T
ax_T.plot(da_scm.T, da_scm.plev/100., ln_scm,
          da_merra2.T, da_merra2.plev/100., ln_merra2,
          da_s11_ctl.T, da_s11_ctl.plev/100., ln_s11_ctl,
          da_s12_ctl.T, da_s12_ctl.plev/100., ln_s12_ctl,
         )
ax_def(ax_T, da_scm.T)

#--- plot Q
ax_Q.plot(da_scm.Q*1000., da_scm.plev/100., ln_scm,
          da_merra2.Q*1000., da_merra2.plev/100., ln_merra2,
          da_s11_ctl.Q*1000., da_s11_ctl.plev/100., ln_s11_ctl,
          da_s12_ctl.Q*1000., da_s12_ctl.plev/100., ln_s12_ctl,
         )
var_dum = da_scm.Q.copy()
var_dum.attrs['units'] = "g/kg"
ax_def(ax_Q, var_dum)
```

## Plot forcing

```python
#--- set default settings of the figures 
def ax_def (ax, var):
    ax.grid(True)
    ax.minorticks_on()
    ax.invert_yaxis()
    ax.legend(["MERRA-2, 3D", 
               "CGILS, S11_ctl, 3D", "CGILS, S11_ctl, hori", "CGILS, S11_ctl, vert", 
               "CGILS, S12_ctl, 3D", "CGILS, S12_ctl, hori", "CGILS, S12_ctl, vert",
              ])
    ax.set_title(var.attrs['long_name'], loc='left')
    ax.set_title(var.attrs['units'], loc='right')
    ax.set_xlabel(var.attrs['long_name']+" ("+var.attrs['units']+")")
    ax.set_ylabel("Pressure (hPa)")

#--- open fig
fig, (ax_divT, ax_divq) = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
fig.suptitle("Forcing", fontsize=20)

#--- set line styles
ln_s11_ctl_h, ln_s11_ctl_v = 'g--', 'g:'
ln_s12_ctl_h, ln_s12_ctl_v = 'y--', 'y:'

#--- plot divT
ax_divT.plot(
          da_merra2.divT3d*86400., da_merra2.plev/100., ln_merra2,
          da_s11_ctl.divT3d*86400., da_s11_ctl.plev/100., ln_s11_ctl,
          da_s11_ctl.divT*86400., da_s11_ctl.plev/100., ln_s11_ctl_h,
          da_s11_ctl.vertdivT*86400., da_s11_ctl.plev/100., ln_s11_ctl_v,
          da_s12_ctl.divT3d*86400., da_s12_ctl.plev/100., ln_s12_ctl,
          da_s12_ctl.divT*86400., da_s12_ctl.plev/100., ln_s12_ctl_h,
          da_s12_ctl.vertdivT*86400., da_s12_ctl.plev/100., ln_s12_ctl_v,
         )
var_dum = da_merra2.divT3d.copy()
var_dum.attrs['units'] = "K/day"
ax_def(ax_divT, var_dum)

#--- plot divq
factor = 864e+5

ax_divq.plot(
          da_merra2.divq3d*factor, da_merra2.plev/100., ln_merra2,
          da_s11_ctl.divq3d*factor, da_s11_ctl.plev/100., ln_s11_ctl,
          da_s11_ctl.divq*factor, da_s11_ctl.plev/100., ln_s11_ctl_h,
          da_s11_ctl.vertdivq*factor, da_s11_ctl.plev/100., ln_s11_ctl_v,
          da_s12_ctl.divq3d*factor, da_s12_ctl.plev/100., ln_s12_ctl,
          da_s12_ctl.divq*factor, da_s12_ctl.plev/100., ln_s12_ctl_h,
          da_s12_ctl.vertdivq*factor, da_s12_ctl.plev/100., ln_s12_ctl_v,
         )
var_dum = da_merra2.divq3d.copy()
var_dum.attrs['units'] = "g/kg/day"
ax_def(ax_divq, var_dum)
```

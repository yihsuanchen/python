```python
#import matplotlib.pyplot as plt
#import numpy as np

#--- 1st dataset
xx1 = np.array([ 93.,  96.,  99., 102., 105., 108., 111., 114., 117.,
                120., 123., 126., 129., 132., 135., 138., 141., 144.,
                147., 150., 153., 156., 159., 162.])
yy1 = np.array([310.7, 308.0, 296.4, 289.5, 288.5, 287.1, 301.1, 308.3,
               311.5, 305.1, 295.6, 292.4, 290.4, 289.1, 299.4, 307.9,
               316.6, 293.9, 291.2, 289.8, 287.1, 285.8, 303.3, 310.])

#--- 2nd dataset
xx2 = np.array([93.,  96.,  99., 102., 105., 108., 111., 114.])
yy2 = np.array([290.4, 289.1, 299.4, 307.9, 316.6, 293.9, 291.2, 289.8])

#--- 3rd dataset
xx3 = np.array([120., 123., 126., 129., 132.])
yy3 = np.array([289.5, 288.5, 287.1, 301.1, 308.3])
```

```python
#######################################
#  Using Matplotlib basic features 
#      https://matplotlib.org/stable/tutorials/introductory/usage.html
#######################################

#--- Create a figure 10 inches by 6 inches. 
fig = plt.figure(figsize=(10, 6))

#--- Add an Axes to the current figure. add_subplot (nrows, ncols, index). 
#      matplotlib.pyplot.subplot, https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html
ax = fig.add_subplot(1, 1, 1)   # or use ax = fig.add_subplot(111) 

#--- Add labels
ax.set_title('test plot')  # title of this figure
ax.set_xlabel('xx')        # x-axis label
ax.set_ylabel('yy')        # y-axis label

#--- plot
linewidth = 3.
ax.plot(xx1, yy1, label='1st dataset', color='red', linestyle='--', linewidth=3.)
ax.plot(xx2, yy2, label='2nd dataset')    # overlay (xx2, yy2)
ax.plot(xx3, yy3, label='3rd dataset', c='orange', ls='-.', lw=5.)
ax.plot(xx3+1, yy3, 'm--')
ax.plot(xx1-10, yy1, 'g--')
ax.plot(xx2+5, yy2, label='6th dataset', c='blue', ls='-.', lw=5.)

#--- draw legend
ax.legend(loc='upper left')    # show da1-da3 & da6

#legend_all = ["da1","da2","da3","da4","da5"]
#ax.legend(legend_all, loc='upper left')  # show da1-da5 datasets

ax.plot(xx2+10, yy2, 'b--', label='7th dataset')  # after ax.legend is called, new legend is not shown


#--- draw grid
ax.grid(True)
```

## XY plot - using pyplot
- Tutorial: https://matplotlib.org/stable/tutorials/introductory/pyplot.html

- pyplot.plot: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot
  - Call signatures:
    - plot([x], y, [fmt], *, data=None, **kwargs)
    - plot([x], y, [fmt], [x2], y2, [fmt2], ..., **kwargs)
  - fmt: string, e.g. 'ro' means 'red dots'
    - Colors: **b, g, r, c, m, y, k, w**. If the color is the only part of the format string, you can additionally use any matplotlib.colors spec, e.g. full names ('green') or hex strings ('#008000'). 
    - Line Styles: **'-', '--', '-.', ':'** {solid, dashed, dash-dot, dotted}
    - Markers: **'o','s','^', ...** {circle, square, triangle_up} 
  - Controlling line properties
    - use keyword arguments, e.g. plt.plot(x, y, linewidth=2.0)
    - use setp, e.g. plt.setp(lines, 'color', 'r', 'linewidth', 2.0)


```python
#######################################
#  Using Matplotlib.pyplot 
#######################################

#------------
#  plot
#-----------

#### default one ####
plt.plot(xx1,yy1)
plt.show()

#### modified one ####
plt.plot(xx1, yy1, 'r--^', xx2, yy2, 'b-s', xx3, yy3, 'g:o')  # plot three datasets with different styles

#--- Add labels
plt.ylabel('yy')
plt.xlabel('xx')
plt.title('Title')

#-- Add texts
plt.text(150., 310., 'aaa')
plt.text(145., 315., '$\mu=100,\ \sigma=15$')

#--- Anootating text
plt.annotate('local max', xy=(120., 310.), xytext=(122., 311.),
             arrowprops=dict(facecolor='black', shrink=0.05),
             )

#--- draw grids
plt.grid(True)

# adjust X & Y range
plt.ylim(270., 320.)
plt.xlim(50., 200.)

#--- show plot
plt.show()
```
